```python
# Define the finite field GF(3)
F = GF(3)
# Create a 4x5 matrix over the field F with rows 1 & 2 & 0 & 1 & 2 \\ 0 & 1 & 2 & 0 & 1 \\ 2 & 0 & 1 & 2 & 0 \\ 1 & 2 & 0 & 1 & 2
M = Matrix(F, [[1, 2, 0, 1, 2],
                [2, 1, 2, 2, 2],
                [2, 1, 0, 2, 1],
                [2, 1, 2, 2, 2]])   

#pretty print M 
pretty_print(M)

M_orig = Matrix(M )

print(latex(M))

# Print the linear transformation f(v_j) for each column v_j of M
for j in range(M.ncols()):
    print("\n f(v_" , j+1 , ") & =&  " , end=' ' )
    for i in range(M.nrows()):
        # The last "+" per j should be removed  
        if i == M.nrows() - 1:
            print (M[i,j] , "\cdot u_" , i + 1, " \\\\", end=' ')
        else:
            print (M[i,j] , "\cdot u_" , i+1 , "+"  , end=' ')



# v is column vector with entries 1,0 , 2,0,1
v = vector(F, [1, 0, 2, 0,1]) 
# Compute f(v) = M * v
f_v = M * v
print("\n f(v) = M * v = ", f_v)    
```

```python
# 4x4 identity matrix over F
Q =  identity_matrix(F, 4)

# Perform Elementary row operation on  and Q
M[1] = M[1] - 2*M[0]
Q[1] = Q[1] - 2*Q[0]

M[2] = M[2] - 2*M[0]
Q[2] = Q[2] - 2*Q[0]

M[3] = M[3] - 2*M[0]
Q[3] = Q[3] - 2*Q[0]



#pretty print M
pretty_print(M) 

print(latex(M)) 

#pretty print Q
pretty_print(Q) 
print(latex(Q))


```

```python
M[3] = M[3] - M[1]
Q[3] = Q[3] - Q[1]

#pretty print M
pretty_print(M) 

print(latex(M)) 

#pretty print Q
pretty_print(Q) 
print(latex(Q))

```

```python
#verify Q * M_orig == M
print("Q * M_orig =")
pretty_print(Q * M_orig)
print("M =")
pretty_print(M)
print("Are they equal?", (Q * M_orig == M))
```
